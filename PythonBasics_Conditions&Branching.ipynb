```python
#Python Basics for beginners
#Conditions and Branching

#Method of equality of two variables using double "Equal Sign", ==, as follows
A=3
B=4
print("A=B: ", A==B)
#the ooutput is False

A=3.6
B=3.6
print("A=B: ",A==B)
#the ooutput is True

#We can measure equality of any types of variables, for example, string
Q1="Beh"
Q2="Beh"
print("Q1=Q2: ",Q1==Q2)

# We can also use "lift operand",<, and "right operand",>,to evaluate (GREATER or SMALLER) two variables
# "lift operand",<
zeta=92.36
beta=99.1
print("Zeta < Beta: ",zeta <beta)
print("Beta < Zeta: ",beta <zeta)

# "right operand",>
x1=106.36
beta=99.1
print("x1 < Beta: ",x1 <beta)
print("Beta < x1: ",beta <x1)

#Greater or equal 
azeta=98.36
abeta=98.1
print("aZeta <= aBeta: ",azeta <abeta)
print("aBeta >= aZeta: ",abeta <azeta)

#Inequality 
#For not equal, you can use "Exclamation Mark",!, with "Equal Sign",=, "!=" as follows
A=3
B=4
print("A!=B: ", A!=B)
#the ooutput is False

A=3.6
B=3.6
print("A!=B: ",A!=B)
#the ooutput is True


```
